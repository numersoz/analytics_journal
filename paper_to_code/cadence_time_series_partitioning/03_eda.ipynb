```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_test = pd.read_parquet("data/processed/data_test.parquet").set_index("dtm")
data_test_2 = pd.read_parquet("data/processed/data_test_2.parquet").set_index("dtm")
data_train = pd.read_parquet("data/processed/data_train.parquet").set_index("dtm")
```

```python
def eda_plots(df):
    # Timeseries:
    fig, ax = plt.subplots(3, 1, figsize = (15, 10), sharex = True)

    df[["light", "co2"]].plot(ax = ax[0], color = ["blue", "orange"])
    ax[0].set_ylabel("Values")
    ax[0].set_title("Light & CO2")

    df[["temp", "humidity"]].plot(ax = ax[1], color = ["green", "purple"])
    ax[1].set_ylabel("Values")
    ax[1].set_title("Temperature & Humidity")

    df["occupancy"].plot(ax = ax[2], drawstyle = "steps-post", color = "red")
    ax[2].set_ylabel("Occupancy")
    ax[2].set_title("Occupancy over time")

    plt.tight_layout()
    plt.show()

    # Correlation Matrix:
    corr_matrix = df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot = True, cmap = "coolwarm", fmt = ".2f", linewidths = 0.5)
    plt.title("Correlation Matrix")
    plt.show()

    # Pair Plot:
    sns.pairplot(df, hue = "occupancy", diag_kind = "kde", plot_kws = {"alpha": 0.6, "s": 80, "edgecolor": "k"})
    plt.show()
```

# Data Train

```python
eda_plots(data_train)
```

# Data Test

```python
eda_plots(data_test)
```

# Data Test 2

```python
eda_plots(data_test_2)
```
